```python
import os
```

```python
languages = ["asm","ben","eng","guj","hin","kan","mal","mar","odi","pun","tam","tel"]
# languages =["asm"]
readFolder = "YT_test"
writeDir = f"combinedData"
for l in languages:
    readDir=f"CSV Files/{l}/{readFolder}"
    writePath = f"{writeDir}/{l}/{readFolder}"
    if os.path.exists(writePath) == False:
        os.makedirs(writePath)
    with open(f"{writePath}/combinedData.csv","a") as f:
        for data in os.listdir(readDir):
            if(data == "combinedData.csv"):
                continue
            with open(f"{os.path.join(readDir,data)}","r",encoding="utf-16") as reader:
                print(f"Appending data from file {data}")
                f.write(reader.read())
        print(f"==== Combined data generated for language {l}")
```
